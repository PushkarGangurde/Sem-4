```python
database = pymysql.connect(
    host='localhost',
    user='root',
    password='Gangurde@156',
    database='demo'
)

cursor = database.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student(
    prn INT PRIMARY KEY,
    name VARCHAR(20),
    subject VARCHAR(20)
)
""")

print("Table 'student' ready")

cursor.close()
database.close()
```

```python
database = pymysql.connect(
    host='localhost',
    user='root',
    password='Gangurde@156',
    database='demo'
)

cursor = database.cursor()

n = int(input("Enter number of records: "))

values = []

for i in range(n):
    print(f"\nEnter details for student {i+1}")
    
    prn = int(input("PRN: "))
    name = input("Name: ")
    subject = input("Subject: ")
    
    values.append((prn, name, subject))

sql = "INSERT INTO student (prn, name, subject) VALUES (%s, %s, %s)"

cursor.executemany(sql, values)

database.commit()

print(cursor.rowcount, "records inserted")

cursor.close()
database.close()
```

```python
database = pymysql.connect(
    host='localhost',
    user='root',
    password='Gangurde@156',
    database='demo'
)

cursor = database.cursor()

prn = int(input("Enter PRN to delete: "))
name = input("Enter Name: ")
subject = input("Enter Subject: ")

sql = "DELETE FROM student WHERE prn=%s AND name=%s AND subject=%s"

cursor.execute(sql, (prn, name, subject))

database.commit()

print(cursor.rowcount, "record deleted")

cursor.close()
database.close()
```
